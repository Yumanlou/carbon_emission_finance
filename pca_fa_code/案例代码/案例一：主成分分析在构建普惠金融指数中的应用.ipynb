```python
#导入数据
import pandas as pd
import numpy as np
X_FI = pd.read_excel(r'C:\Users\LENOVO\第二章\案例数据\FI.xlsx')
X_FI.head()
```

```python
# 进行标准化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler() 
scaler.fit(X_FI)  
X2_FI = pd.DataFrame(scaler.transform(X_FI))  
X2_FI.head()  # 查看标准化后的数据列
```

```python
# PCA
import matplotlib.pyplot as plt  
from sklearn.decomposition import PCA  
pca = PCA(n_components=X2_FI.shape[1]-1)  # 创建主成分分析对象，设定主成分数为样本特征数 - 1
reduced_x = pca.fit_transform(X2_FI)  # 基于已标准化数据创建主成分分析模型
covper = pca.explained_variance_  # 降维后的各主成分的方差值
covper = pd.DataFrame(np.round(covper, 3))  
covper.to_excel('covper.xlsx')

# 绘制碎石图
plt.plot(covper, 'bx--')
plt.xlabel('Component #n')
plt.ylabel('Variance')
plt.show()
```

```python
coefficient=pd.DataFrame(np.round(pca.components_, 3),columns=X_FI.columns)#输出载荷系数
coefficient.to_excel('coefficient.xlsx')
```

```python
contribution=pd.DataFrame(np.round(pca.explained_variance_ratio_, 3) )#输出累计方差贡献率
contribution.to_excel('contribution.xlsx')
```
